# Univerzitní města

Chceme zjistit, ve kterém státě je nejvíce univerzit. A ve kterém městě.

```python
# Proměnná, kde bude aktuální stát
current_state = ""
output = []
with open("university_towns.txt", encoding="utf-8") as file:
    for line in file:
        # Chci se podívat, zda se mění aktuální
        if "edit" in line:
            current_state = line.replace("[edit]", "")
            current_state = current_state.strip()
        else:
            # Kde je hranatá závorka?
            br_pos = line.find("[")
            # Našlo to hranatou závorku?
            if br_pos >= 0:
                # Nechám si jen znaky ke hranaté závorce
                line = line[:br_pos]
            line_list = line.split("(")
            city = line_list[0].strip()
            universities = line_list[1:]
            # Chceme získat seznam škol pro každé město
            universities_list = []
            # append -> nested
            # [["A", "B"], ["C", "D"]]
            # extend -> flat
            # ["A", "B", "C", "D"]
            for item in universities:
                # Nejprve se zbavíme konce řádků a poté zbytečných závorek
                item = item.strip().strip(")")
                # Chceme získat univerzity samostatně, řetězec rozdělíme na prvky seznamu
                item_split = item.split(", ")
                # Přidáme univerzity do universities_list
                # Toto je zkrácený zápis řádku níže
                # universities_list += item_split
                universities_list = universities_list + item_split
                # Alternativně můžeme použít metodu extend
                # universities_list.extend(item_split)
            for university in universities_list:
                # Přidáváme vnořený seznam se strukturou [stát, město, univerzita]
                output.append([current_state, city, university])
with open("universities.csv", "w", encoding="utf-8") as file:
    for line in output:
        # [stát, město, univerzita] -> stát,město,univerzita
        # stát
        # město
        # univerzita
        print(",".join(line), file=file)
```

Nyní si můžeme zkusit spočítat počet univerzit v jednoltivých státech s využitím Pythonu (`Counter`) a porovnáme to s použitím pandas.

```python
from collections import Counter

counter = Counter()
for item in output:
    key = item[0]
    counter[key] += 1
print(counter)
```

```python
import pandas

# Ze svého dvourozměrného seznamu si vytvořím pandas DataFrame. Stačí vložit dvourozměrný seznam a pandas si s ním poradí.
data = pandas.DataFrame(output)
# Protože u dvourozměrného seznamu nemáme názvy sloupců, nastavíme si je ručně
data.columns = ["state", "town", "university"]
# Agregace
data_grouped = data.groupby("state").size()
# Výpis výsledků - řadíme od nejvyššího
data_grouped.sort_values(ascending=False)
```

Můžeme najít i město s největším počtem univerzit.

```python
data_grouped = data.groupby(["state", "town"]).size()
data_grouped.sort_values(ascending=False)
```

V Pythonu načteme data pomocí cyklu a délku zjistíme s využitím funkce `len()`.

Všimni si, že hodnota je o 1 vyšší, protože v Pythonu počítáme záhlaví tabulky jako řádek.

```python
data_list = []
with open("Flickr-Images-Book.tsv", encoding="utf-8") as file:
    for line in file:
        line = line.split("\t")
        data_list.append(line)
len(data_list)
```

Je `Identifier` unikátní? V pandas můžeme použít vlastnost `is_unique`.

```python
data["Identifier"].is_unique
```

V Pythonu můžeme použít množinu, která má pouze unikátní hodnoty. Porovnáme délku množiny a seznamu a podle toho vidíme, zda se nám tam nějaký prvek opakuje.

```python
index_identifier = 0

identifiers = []
for item in data_list[1:]:
    identifiers.append(item[index_identifier])
identifiers_set = set(identifiers)
if len(identifiers) == len(identifiers_set):
    print("Jsou unikátní")
else:
    print("Nejsou unikátní")
```
